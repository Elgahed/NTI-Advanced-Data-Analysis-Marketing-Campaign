```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import pickle
```

```python
def load_and_explore_data_simple():

    print("Using sample data as no file was uploaded.")

    return pd.DataFrame()

def load_data_with_upload():
    print("=" * 50)
    print("MANUAL FILE UPLOAD")
    print("=" * 50)

    try:
        from google.colab import files
        uploaded = files.upload()

        import io
        excel_files = [f for f in uploaded.keys() if f.endswith(('.xlsx', '.xls'))]

        if excel_files:
            file_name = excel_files[0]
            df = pd.read_excel(io.BytesIO(uploaded[file_name]),
                              sheet_name='marketing_campaign_dataset')
            print(f"✅ Successfully loaded: {file_name}")
        else:
            print("❌ No Excel file found. Using sample data.")
            df = load_and_explore_data_simple()

    except Exception as e:
        print(f"❌ Upload error: {e}")
        df = load_and_explore_data_simple()

    return df

print("Choose loading method:")
print("1 - Upload file manually")
print("2 - Use sample data")
choice = input("Enter choice (1 or 2): ")

if choice == "1":
    df = load_data_with_upload()
else:
    df = load_and_explore_data_simple()
```

```python
print("=" * 50)
print("DATA EXPLORATION")
print("=" * 50)

print(f"Dataset Shape: {df.shape}")
print(f"\nColumns: {list(df.columns)}")
print(f"\nData Types:\n{df.dtypes}")
print(f"\nMissing Values:\n{df.isnull().sum()}")
print(f"\nFirst 3 rows:\n{df.head(3)}")

print(f"\nDescriptive Statistics:\n{df.describe()}")
```

```python
def clean_data(df):
    print("\n" + "=" * 50)
    print("DATA CLEANING")
    print("=" * 50)

    df_clean = df.copy()

    initial_rows = len(df_clean)
    df_clean = df_clean.dropna()
    removed_rows = initial_rows - len(df_clean)
    print(f"Removed {removed_rows} rows with missing values")

    df_clean['Date'] = pd.to_datetime(df_clean['Date'])
    df_clean['Year'] = df_clean['Date'].dt.year
    df_clean['Month'] = df_clean['Date'].dt.month
    df_clean['Day'] = df_clean['Date'].dt.day
    df_clean['DayOfWeek'] = df_clean['Date'].dt.dayofweek
    df_clean = df_clean.drop(columns=['Date'])
    print("Date features extracted and original date column dropped")

    return df_clean

df_clean = clean_data(df)
```

```python
def preprocess_features(df_clean):
    print("\n" + "=" * 50)
    print("FEATURE PREPROCESSING")
    print("=" * 50)

    df_processed = df_clean.copy()

    categorical_cols = ['Campaign', 'City', 'Channel', 'Device', 'Ad_Type']
    label_encoders = {}

    for col in categorical_cols:
        if col in df_processed.columns:
            le = LabelEncoder()
            df_processed[col] = le.fit_transform(df_processed[col].astype(str))
            label_encoders[col] = le
            print(f"Encoded categorical column: {col}")

    numeric_cols = ['Latitude', 'Longitude', 'Impressions', 'CTR', 'Clicks',
                   'Avg_CPC', 'Spend_GBP', 'Conversions', 'Conversion_Value_GBP',
                   'Likes', 'Shares', 'Comments', 'Conversion_Rate', 'Cost per click',
                   'ROI', 'ROAS', 'Cost per 1000 Impressions', 'CPA', 'Profit']

    for col in numeric_cols:
        if col in df_processed.columns:
            df_processed[col] = pd.to_numeric(df_processed[col], errors='coerce')

    df_processed = df_processed.dropna()
    print(f"Final dataset shape after preprocessing: {df_processed.shape}")

    return df_processed, label_encoders

df_processed, label_encoders = preprocess_features(df_clean)
```

```python
def prepare_train_test_data(df_processed, target_column='Conversion_Value_GBP'):
    print("\n" + "=" * 50)
    print("TRAIN-TEST SPLIT PREPARATION")
    print("=" * 50)


    feature_columns = [col for col in df_processed.columns if col != target_column]
    print(f"Using {len(feature_columns)} features")

    X = df_processed[feature_columns]
    y = df_processed[target_column]


    numerical_cols_to_scale = X.select_dtypes(include=[np.number]).columns.tolist()

    scaler = StandardScaler()
    X[numerical_cols_to_scale] = scaler.fit_transform(X[numerical_cols_to_scale])
    print(f"Scaled {len(numerical_cols_to_scale)} numerical features")

    return X, y, feature_columns, scaler

X, y, feature_columns, scaler = prepare_train_test_data(df_processed)
```

```python
def split_data(X, y, test_size=0.2, val_size=0.2, random_state=42):
    print("\n" + "=" * 50)
    print("DATA SPLITTING")
    print("=" * 50)

    # Split into train+val and test
    X_train_val, X_test, y_train_val, y_test = train_test_split(
        X, y, test_size=test_size, random_state=random_state, shuffle=True
    )

    # Split train_val into train and validation
    X_train, X_val, y_train, y_val = train_test_split(
        X_train_val, y_train_val, test_size=val_size, random_state=random_state
    )

    print(f"Total samples: {len(X)}")
    print(f"Training set: {X_train.shape[0]} samples ({X_train.shape[0]/len(X)*100:.1f}%)")
    print(f"Validation set: {X_val.shape[0]} samples ({X_val.shape[0]/len(X)*100:.1f}%)")
    print(f"Test set: {X_test.shape[0]} samples ({X_test.shape[0]/len(X)*100:.1f}%)")
    print(f"Number of features: {X_train.shape[1]}")

    return X_train, X_val, X_test, y_train, y_val, y_test

X_train, X_val, X_test, y_train, y_val, y_test = split_data(X, y)
```

```python
def save_processed_data(X_train, X_val, X_test, y_train, y_val, y_test,
                       feature_columns, label_encoders, scaler, file_path):
    processed_data = {
        'X_train': X_train,
        'X_val': X_val,
        'X_test': X_test,
        'y_train': y_train,
        'y_val': y_val,
        'y_test': y_test,
        'feature_columns': feature_columns,
        'label_encoders': label_encoders,
        'scaler': scaler
    }

    with open(file_path, 'wb') as f:
        pickle.dump(processed_data, f)

    print(f"\nProcessed data saved to {file_path}")
    return processed_data

processed_data = save_processed_data(
    X_train, X_val, X_test, y_train, y_val, y_test,
    feature_columns, label_encoders, scaler, 'marketing_data_processed.pkl'
)
```

```python
def train_and_evaluate_model(X_train, y_train, X_val, y_val, X_test, y_test):
    print("\n" + "=" * 50)
    print("MODEL TRAINING AND EVALUATION")
    print("=" * 50)

    model = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
    model.fit(X_train, y_train)
    print("Model training completed")

    y_pred_train = model.predict(X_train)
    y_pred_val = model.predict(X_val)
    y_pred_test = model.predict(X_test)

    def evaluate_model(y_true, y_pred, set_name):
        mse = mean_squared_error(y_true, y_pred)
        rmse = np.sqrt(mse)
        mae = mean_absolute_error(y_true, y_pred)
        r2 = r2_score(y_true, y_pred)

        print(f"\n{set_name} Set Evaluation:")
        print(f"MSE: {mse:.2f}")
        print(f"RMSE: {rmse:.2f}")
        print(f"MAE: {mae:.2f}")
        print(f"R² Score: {r2:.4f}")

        return rmse, r2

    evaluate_model(y_train, y_pred_train, "Training")
    evaluate_model(y_val, y_pred_val, "Validation")
    evaluate_model(y_test, y_pred_test, "Test")

    return model

model = train_and_evaluate_model(X_train, y_train, X_val, y_val, X_test, y_test)
```

```python
def visualize_data(df, target_column='Conversion_Value_GBP'):
    print("=" * 50)
    print("DATA VISUALIZATION")
    print("=" * 50)

    df_viz = df.copy()

    plt.figure(figsize=(15, 12))

    plt.subplot(2, 3, 1)
    plt.hist(df_viz[target_column], bins=30, alpha=0.7, color='skyblue', edgecolor='black')
    plt.title(f'Distribution of {target_column}')
    plt.xlabel(target_column)
    plt.ylabel('Frequency')
    plt.grid(True, alpha=0.3)

    plt.subplot(2, 3, 2)
    plt.scatter(df_viz['Spend_GBP'], df_viz[target_column], alpha=0.6, color='orange')
    plt.title('Spend vs Conversion Value')
    plt.xlabel('Spend (GBP)')
    plt.ylabel('Conversion Value (GBP)')
    plt.grid(True, alpha=0.3)

    plt.subplot(2, 3, 3)
    plt.scatter(df_viz['Clicks'], df_viz[target_column], alpha=0.6, color='green')
    plt.title('Clicks vs Conversion Value')
    plt.xlabel('Clicks')
    plt.ylabel('Conversion Value (GBP)')
    plt.grid(True, alpha=0.3)

    plt.subplot(2, 3, 4)
    campaign_stats = df_viz.groupby('Campaign')[target_column].mean()
    campaign_stats.plot(kind='bar', color=['red', 'green', 'blue'], alpha=0.7)
    plt.title('Average Conversion Value by Campaign')
    plt.xlabel('Campaign')
    plt.ylabel('Average Conversion Value (GBP)')
    plt.xticks(rotation=45)
    plt.grid(True, alpha=0.3)

    if 'Date' in df_viz.columns:
        plt.subplot(2, 3, 5)
        df_viz['Date'] = pd.to_datetime(df_viz['Date'])
        daily_values = df_viz.groupby(df_viz['Date'].dt.date)[target_column].sum()
        daily_values.plot(kind='line', color='purple', marker='o')
        plt.title('Daily Conversion Value Over Time')
        plt.xlabel('Date')
        plt.ylabel('Conversion Value (GBP)')
        plt.grid(True, alpha=0.3)

    plt.subplot(2, 3, 6)
    plt.scatter(df_viz['CTR'], df_viz[target_column], alpha=0.6, color='brown')
    plt.title('CTR vs Conversion Value')
    plt.xlabel('CTR')
    plt.ylabel('Conversion Value (GBP)')
    plt.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

    print("\n📈 Correlation Matrix:")
    numeric_cols = df_viz.select_dtypes(include=[np.number]).columns
    correlation_matrix = df_viz[numeric_cols].corr()

    plt.figure(figsize=(12, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0,
                fmt='.2f', linewidths=0.5)
    plt.title('Correlation Matrix of Numerical Features')
    plt.show()

    plt.figure(figsize=(15, 5))

    plt.subplot(1, 3, 1)
    scatter = plt.scatter(df_viz['Spend_GBP'], df_viz['Clicks'],
                         c=df_viz[target_column], cmap='viridis', alpha=0.7)
    plt.colorbar(scatter, label=target_column)
    plt.title('Spend vs Clicks (Color: Conversion Value)')
    plt.xlabel('Spend (GBP)')
    plt.ylabel('Clicks')
    plt.grid(True, alpha=0.3)

    plt.subplot(1, 3, 2)
    if 'Device' in df_viz.columns:
        device_stats = df_viz.groupby('Device')[target_column].mean()
        device_stats.plot(kind='bar', color=['lightblue', 'lightgreen'], alpha=0.7)
        plt.title('Conversion Value by Device')
        plt.xlabel('Device')
        plt.ylabel('Average Conversion Value (GBP)')
        plt.xticks(rotation=0)
        plt.grid(True, alpha=0.3)

    plt.subplot(1, 3, 3)
    if 'Channel' in df_viz.columns:
        channel_stats = df_viz.groupby('Channel')[target_column].mean()
        channel_stats.plot(kind='bar', color='orange', alpha=0.7)
        plt.title('Conversion Value by Channel')
        plt.xlabel('Channel')
        plt.ylabel('Average Conversion Value (GBP)')
        plt.xticks(rotation=0)
        plt.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(15, 5))

    numerical_features = ['Spend_GBP', 'Clicks', 'Impressions', target_column]

    for i, feature in enumerate(numerical_features, 1):
        if feature in df_viz.columns:
            plt.subplot(1, 4, i)
            plt.boxplot(df_viz[feature].dropna())
            plt.title(f'Boxplot of {feature}')
            plt.ylabel(feature)

    plt.tight_layout()
    plt.show()

    if 'Date' in df_viz.columns:
        plt.figure(figsize=(15, 10))

        df_viz['Month'] = df_viz['Date'].dt.month
        monthly_stats = df_viz.groupby('Month')[target_column].agg(['mean', 'sum', 'count'])

        plt.subplot(2, 2, 1)
        monthly_stats['mean'].plot(kind='bar', color='teal', alpha=0.7)
        plt.title('Monthly Average Conversion Value')
        plt.xlabel('Month')
        plt.ylabel('Average Value (GBP)')
        plt.grid(True, alpha=0.3)

        plt.subplot(2, 2, 2)
        monthly_stats['sum'].plot(kind='bar', color='coral', alpha=0.7)
        plt.title('Monthly Total Conversion Value')
        plt.xlabel('Month')
        plt.ylabel('Total Value (GBP)')
        plt.grid(True, alpha=0.3)

        df_viz['DayOfWeek'] = df_viz['Date'].dt.dayofweek
        daily_stats = df_viz.groupby('DayOfWeek')[target_column].mean()

        plt.subplot(2, 2, 3)
        daily_stats.plot(kind='bar', color='purple', alpha=0.7)
        plt.title('Conversion Value by Day of Week')
        plt.xlabel('Day of Week (0=Monday, 6=Sunday)')
        plt.ylabel('Average Value (GBP)')
        plt.grid(True, alpha=0.3)

        plt.subplot(2, 2, 4)
        for campaign in df_viz['Campaign'].unique():
            campaign_data = df_viz[df_viz['Campaign'] == campaign]
            monthly_campaign = campaign_data.groupby('Month')[target_column].mean()
            plt.plot(monthly_campaign.index, monthly_campaign.values,
                    marker='o', label=campaign, linewidth=2)

        plt.title('Monthly Trends by Campaign')
        plt.xlabel('Month')
        plt.ylabel('Average Conversion Value (GBP)')
        plt.legend()
        plt.grid(True, alpha=0.3)

        plt.tight_layout()
        plt.show()

    print("✅ All visualizations completed successfully!")

visualize_data(df)
```
